```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('seaborn')
%matplotlib inline
```

$$m_i= ∑_j l_j · Ax_{i,j}$$

$$l_{k,j} = (1 − j/J) − (k/d)(1 − 2j/J)$$

```python
def l_kj(k, j, J=J, d=d):
    return (1 - (j + 1) / J) - ((k + 1) / d) * (1 - 2 * (j + 1) / J)
```

```python
def draw_pe(sentence):
    # Set J, d
    J = len(sentence.split())
    d = 128
    
    L = np.zeros(shape=(d, J))
    for i in range(d):
        for j in range(J):
            L[i,j] = l_kj(i, j)
            
    # draw plot
    fig, ax = plt.subplots(figsize=(25, 20))
    ax.imshow(L.T, cmap='magma_r')
    plt.xlabel('dim', size='14')
    plt.ylabel('poisition', size='14')
```

```python
sentence = 'The same representation is used for questions, memory inputs and memory outputs.'
draw_pe(sentence)
```

```python
sentence = 'We therefore propose a second representation that encodes the position of words within the sentence.'
draw_pe(sentence)
```
