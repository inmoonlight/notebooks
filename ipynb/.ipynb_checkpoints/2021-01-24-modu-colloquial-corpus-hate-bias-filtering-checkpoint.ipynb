## Set threshold

#### RE
- hate/offensive: 0.9
- bias: 0.5

#### NLI
- hate/offensive: 0.46
- bias: 0.99

```python
spoken
```

```python
def filter_hate(df, label, thres):
    return df[(df['hate'] == label) & (df['hate_score'] >= thres)].sort_values('hate_score', ascending=False)
```

```python
def filter_bias(df, thres):
    return df[(df['gender_bias'] == 'True')].sort_values('gender_bias_score', ascending=False)
```

```python
def filter_hate_bias(df, hate_label, hate_thres, gender_bias_thres):
    return df[(df['hate'] == hate_label) & (df['hate_score'] >= hate_thres) & (df['gender_bias'] == 'True')].sort_values('hate_score', ascending=False)
#     return df[(df['hate'] == hate_label) & (df['hate_score'] >= hate_thres) & (df['gender_bias_score'] >= gender_bias_thres)].sort_values('hate_score', ascending=False).sort_values('gender_bias_score', ascending=False)
```

```python
hate = 0.9
bias = 0.99
```

#### spoken

```python
filter_hate_bias(spoken, 'hate', 0.90, bias)
```

```python
filter_hate(spoken, 'hate', hate).head(20)
```

```python
filter_bias(spoken, bias).head(30)
```

#### web

```python
filter_hate_bias(web, 'hate', 0.90, bias)
```

```python
filter_hate(web, 'hate', hate).head(50)
```

```python
filter_bias(web, bias).tail(50)
```
